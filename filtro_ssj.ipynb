```python
import cv2
import FaceDetectors
import numpy as np
import pygame  

# --- INICIALIZAR SONIDO ---
pygame.mixer.init()
sonido_ssj = pygame.mixer.Sound("sonido_ssj.mp3") 
sonido_ssj.set_volume(0.5) 
sonido_activo = False 

# --- FUNCIN NICA DE MEZCLA ---
def overlay_transparent(background, overlay, x, y, overlay_size=None):
    if overlay_size is not None:
        overlay = cv2.resize(overlay.copy(), overlay_size)
        
    bg_h, bg_w, _ = background.shape
    img_h, img_w, c = overlay.shape

    if x + img_w > bg_w: img_w = bg_w - x
    if y + img_h > bg_h: img_h = bg_h - y
    if x < 0: img_w = img_w + x; x = 0
    if y < 0: img_h = img_h + y; y = 0
    if img_w <= 0 or img_h <= 0: return background

    crop_fg = overlay[:img_h, :img_w]
    crop_bg = background[y:y+img_h, x:x+img_w]
    
    alpha = crop_fg[:, :, 3] / 255.0
    alpha_inv = 1.0 - alpha
    
    for c in range(0, 3):
        crop_bg[:, :, c] = (alpha * crop_fg[:, :, c] + alpha_inv * crop_bg[:, :, c])
    return background

# --- CARGA ---
img_pelo = cv2.imread('peluca_ssj.png', cv2.IMREAD_UNCHANGED)
img_aura = cv2.imread('aura_ki.png', cv2.IMREAD_UNCHANGED)

# --- CONFIGURACIN ---
FDet = FaceDetectors.FaceDetector()
cap = cv2.VideoCapture(0)
cap.set(3, 640); cap.set(4, 480)

while True:
    ret, frame = cap.read()
    if ret:
        values = FDet.SingleFaceEyesDetection(frame, FDet.FaceDetectors[1], FDet.EyeDetectors[1])
        gritando_ahora = False

        if values and values[2] is not None and len(values[2]) >= 68:
            face, eyes, shape = values
            
            apertura = shape[66][1] - shape[62][1]      
            umbral = (shape[8][1] - shape[27][1]) * 0.10 

            if apertura > umbral:
                gritando_ahora = True 
                
                # --- LOGICA DE SONIDO ---
                if not sonido_activo and sonido_ssj:
                    sonido_ssj.play()
                    sonido_activo = True

                # --- TEMBLOR ---
                intensidad = 15
                shift_x = np.random.randint(-intensidad, intensidad)
                shift_y = np.random.randint(-intensidad, intensidad)
                M = np.float32([[1, 0, shift_x], [0, 1, shift_y]])
                frame = cv2.warpAffine(frame, M, (frame.shape[1], frame.shape[0]))

                # --- DIBUJOS ---
                ancho_cara = shape[16][0] - shape[0][0]
                
                # Aura
                ancho_aura = int(ancho_cara * 3.5)
                factor_a = ancho_aura / img_aura.shape[1]
                x_a = shape[8][0] - int(ancho_aura / 2)
                y_a = shape[8][1] - int(img_aura.shape[0] * factor_a / 1.5)
                try: frame = overlay_transparent(frame, img_aura, x_a, y_a, (ancho_aura, int(img_aura.shape[0] * factor_a)))
                except: pass

                # Pelo
                ancho_pelo = int(ancho_cara * 1.8)
                factor_p = ancho_pelo / img_pelo.shape[1]
                x_p = shape[27][0] - int(ancho_pelo / 2)
                y_p = shape[27][1] - int(img_pelo.shape[0] * factor_p * 0.85)
                try: frame = overlay_transparent(frame, img_pelo, x_p, y_p, (ancho_pelo, int(img_pelo.shape[0] * factor_p)))
                except: pass
        
        # --- RESET DEL SONIDO ---
        if not gritando_ahora and sonido_activo:
             if sonido_ssj: sonido_ssj.stop() 
             sonido_activo = False

        cv2.imshow('Super Saiyan Audio', frame)
        if cv2.waitKey(1) == 27: break 

cap.release()
cv2.destroyAllWindows()
pygame.quit() 
```
